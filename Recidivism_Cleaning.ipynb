```python
# Dependencies
import pandas as pd
```

```python
# Reading in CSV
df = pd.read_csv("Resources/Recidivism.csv")
```

```python
df.head()
```

```python
# Dropping unnecessary columns
df = df.drop(columns=['Recidivism Reporting Year', 'Convicting Offense Subtype', 'Main Supervising District', 'Release type: Paroled to Detainder united', 'Part of Target Population'], axis=1)
df.head()
```

```python
# Simplifying race/ethnic categories
df.loc[(df["Race - Ethnicity"] == "White - Hispanic"), "Race - Ethnicity"] = "Hispanic"
```

```python
df.loc[(df["Race - Ethnicity"] == "White - Non-Hispanic"), "Race - Ethnicity"] = "White"
```

```python
df.loc[(df["Race - Ethnicity"] == "Black - Hispanic"), "Race - Ethnicity"] = "Hispanic"
```

```python
df.loc[(df["Race - Ethnicity"] == "Black - Non-Hispanic"), "Race - Ethnicity"] = "Black"
```

```python
df.loc[(df["Race - Ethnicity"] == "Asian or Pacific Islander - Hispanic"), "Race - Ethnicity"] = "Hispanic"
```

```python
df.loc[(df["Race - Ethnicity"] == "Asian or Pacific Islander - Non-Hispanic"), "Race - Ethnicity"] = "Asian"
```

```python
df.loc[(df["Race - Ethnicity"] == "American Indian or Alaska Native - Hispanic"), "Race - Ethnicity"] = "Hispanic"
```

```python
df.loc[(df["Race - Ethnicity"] == "American Indian or Alaska Native - Non-Hispanic"), "Race - Ethnicity"] = "Native"
```

```python
df.loc[(df["Race - Ethnicity"] == "White -"), "Race - Ethnicity"] = "White"
```

```python
df.loc[(df["Race - Ethnicity"] == "Black -"), "Race - Ethnicity"] = "Black"
```

```python
df.loc[(df["Race - Ethnicity"] == "N/A -"), "Race - Ethnicity"] = "Other"
```

```python
df.loc[(df["Race - Ethnicity"] == ""), "Race - Ethnicity"] = "Other"
df
```

```python
df["Race - Ethnicity"].nunique()
```

```python
# Replacing blanks in "Age At Release" with "N/A"
df.loc[(df["Age At Release "] == ""), "Age At Release "] = "N/A"
```

```python
df = df.rename({"Fiscal Year Released": "Year Released", "Race - Ethnicity": "Race", "Age At Release ": "Age", "Convicting Offense Classification": "Offense Classification", "Convicting Offense Type": "Offense Type", "Recidivism - Return to Prison numeric": "Recidivism"}, axis=1)
```

```python
# Replacing categorical variables with dummies
dummies_df = pd.get_dummies(df.drop(columns=["Year Released", "Recidivism"], axis=1))
dummies_df
```

```python
dummies_df["Year Released"] = df["Year Released"]
dummies_df
```

```python
dummies_df["Recidivism"] = df["Recidivism"]
dummies_df
```

```python
# Exporting CSV of new dataframe
dummies_df.to_csv("Resources/Recidivism_Cleaned.csv")
```
